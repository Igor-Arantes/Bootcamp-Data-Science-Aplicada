# Importação e exploração dos Dados

```python
#Descompactando e Importando
zip= ZipFile('Dados/Dados.zip')
dados = pd.read_excel(zip.open(zip.namelist()[0]))
dados.sample(5)
```

```python
#Entendendo o Df
print(dados.describe())
print('-------')
print(dados.shape)
print('-------')
print(dados.info)
print('-------')
print(dados.sample(5))
```

```python
#Foi pra UTI ou não
dados['ICU'].value_counts()
```

```python
#Remover Nan e Infinito
dados_limpos = dados.dropna()
```

```python
#Separando os dados para o modelo
y = dados_limpos["ICU"] # y = resultado
x = dados_limpos.drop(["ICU"],axis=1) #x = entrada
```

```python
# Removendo dados não float
x_columns = dados_limpos.describe().columns 
x = dados_limpos[x_columns].drop(["ICU"],axis=1) #x = entrada
```

```python
# Criando objeto modelo e passando os dados
modelo = LogisticRegression(max_iter=560)
modelo.fit(x,y)
```

```python
modelo.predict(x)
```
