Apenas alterando as esclas e o  zoom do gráfico é possivel passar uma  informação, não necessariamente falsa. Porem mostra um apenas um ponto de  vista da realidade dando uma falsa impressão do que realmente acontece. # Desafio: gráfico de barra do gasto dos outros estados em função ao seu estado
