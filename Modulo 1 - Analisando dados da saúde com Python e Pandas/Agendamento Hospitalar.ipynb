```python
idh= pd.read_excel('Dados/idhmbrasil.xlsx')
idh = idh.iloc[1:-3]
```

```python
ordenado_por_total[['Codigo','Unidade da Federação']] =ordenado_por_total.index.str.split(n=1,expand=True).tolist()
ordenado_por_total.set_index('Codigo',inplace=True)
order=ordenado_por_total.columns[144:146].to_list()
for i in range(len(ordenado_por_total.columns[:-2])):
    order.append(ordenado_por_total.columns[i])
ordenado_por_total=ordenado_por_total.reindex(columns=order)    
```

```python
meses_interesse = ordenado_por_total.columns.to_list()[0:1]
meses_interesse.extend(ordenado_por_total.columns.to_list()[131:-1])
gasto_idh = ordenado_por_total[meses_interesse]
gasto_idh['media']=gasto_idh.mean(axis=1)
gasto_idh['media']=gasto_idh['media']*1000000
gasto_idh=gasto_idh.sort_values(by='Unidade da Federação')
gasto_idh['IDHM'] = idh.sort_values(by='Territorialidades')['IDHM 2017'].to_list()
gasto_idh['Populacao'] = idh.sort_values(by='Territorialidades')['População total 2010'].to_list()
gasto_idh['Gasto por Habitante']=round(gasto_idh['media']/gasto_idh['Populacao'],2)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(2,1,1)
ax=sns.barplot(data=gasto_idh,
            y=gasto_idh['Gasto por Habitante'],
            x=gasto_idh['Unidade da Federação'],color='b')
plt.xticks(color='k',rotation=80)
ax.set_xlabel('')
plt.grid()
plt.subplot(2,1,2)
ax=sns.barplot(data=gasto_idh,
            y=gasto_idh['IDHM'],
            x=gasto_idh['Unidade da Federação'],color='b')
plt.xticks(color='w',rotation=80)
plt.ylim(0,1)
plt.grid()


plt.show()
```
