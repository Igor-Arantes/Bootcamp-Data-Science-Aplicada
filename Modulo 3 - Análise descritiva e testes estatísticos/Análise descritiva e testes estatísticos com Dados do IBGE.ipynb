Desafio 3: Criar uma funçao para gerar a tabela de frequencia e analisar outras informações

```python
def tabela_frequencia(dataframe,coluna,pergunta):
    frequencia = dataframe[coluna].value_counts()
    percentual = dataframe[coluna].value_counts(normalize=True)*100
    dist_frequ = pd.DataFrame({'Frequência': frequencia, 'Percentual': percentual})
    dist_frequ.rename_axis(pergunta, axis = 1, inplace=True)
    return dist_frequ
    
```

```python
tabela_frequencia(dados,'VB11007','Sentimento em relação ao seu corpo')
```

Desafio 7: Calcular o desvio Padrão, Media, Moda e Mediana e plotar no grafico

```python
desvio_padrao = dados['VB17004'].std()
media = dados['VB17004'].mean()
moda = dados['VB17004'].mode()[0]
mediana = dados['VB17004'].median()

print("Desvio Padrao: {} Media: {}, Moda: {}, Mediana: {}".format(round(desvio_padrao,2),round(media,2),moda,mediana))
```

```python
ax = sns.histplot(data=dados,x='VB17004',color='blue', kde=True)
ax.figure.set_size_inches(14,8)
ax.set_title("Distribuição de frequencia")
ax.set_xlabel('Altura')
ax.axvline(media,c='y',label='Media {}'.format(round(media,2)))
ax.axvline(moda,c='g',label='Moda {}'.format(moda))
ax.axvline(mediana,c='r',label='Mediana{}'.format(mediana))
ax.annotate('Dp = {}'.format(round(desvio_padrao,2)),xy=(190,600),fontsize=15)
ax.legend()
ax;
```
