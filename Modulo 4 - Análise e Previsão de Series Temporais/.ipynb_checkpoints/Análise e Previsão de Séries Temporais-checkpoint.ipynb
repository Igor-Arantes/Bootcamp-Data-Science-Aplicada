## Desafio-1 - Transformar dados em uma serie ano-mes

```python
dados_st = dados.melt(id_vars='Ano Diagnóstico', value_name='Casos', var_name='Mês')
```

```python
dados_st['Data'] = dados_st['Ano Diagnóstico']+'-'+dados_st['Mês']
dados_st
```

Fim do Desafio

```python
plt.figure(figsize=(18,12))
sns.lineplot(x='Data', y='Casos', data=dados_st[:30])
```

## Desafio-2 - Converter a data para Datetime

```python
#pd.to_datetime(dados_st['Data'],yearfirst=True)

#Erro pois os nomes estao em portugues
```

```python
dic_mes = {
    'Jan':'Jan',
    'Fev':'Feb',
    'Mar':'Mar',
    'Abr':'Apr',
    'Mai':'May',
    'Jun':'Jun',
    'Jul':'Jul',
    'Ago':'Aug',
    'Set':'Sep',
    'Out':'Oct',
    'Nov':'Nov',
    'Dez':'Dec',
}
dados_st['Mês'] = dados_st['Mês'].map(dic_mes)
dados_st['Data'] = dados_st['Ano Diagnóstico']+'-'+dados_st['Mês']
dados_st['Data'] = pd.to_datetime(dados_st['Data'])
```

```python
dados_st
```

 Fim do Desafio

```python
dados_st = dados_st[['Data','Casos']]
dados_st
```

```python
plt.figure(figsize=(22,12))
sns.lineplot(x='Data', y='Casos', data=dados_st)
plt.xticks(rotation=90)
plt.show()
```

## Desafio 03 - Realizar os mesmos processos de transformação de Série temporal para base de dados de Hanseniase

```python
dados_hanseniase=pd.read_csv("Dados/hanseaniase_ano_mes_diagnostico.csv",encoding = "ISO-8859-1",sep=';', skipfooter=16, engine="python")
dados_hanseniase.drop('Total', axis=1, inplace=True)
```

```python
dados_hanseniase_ano_mes = dados_hanseniase.melt(id_vars='Ano Diagnóstico', value_name='Casos', var_name='Mês')
dados_hanseniase_ano_mes['Ano Diagnóstico'] = dados_hanseniase_ano_mes['Ano Diagnóstico'].apply(str)
```

```python
dados_hanseniase_ano_mes['Mês'] = dados_hanseniase_ano_mes['Mês'].map(dic_mes)
dados_hanseniase_ano_mes['Data'] = dados_hanseniase_ano_mes['Ano Diagnóstico']+ "-" +dados_hanseniase_ano_mes['Mês']
dados_hanseniase_ano_mes['Data'] = pd.to_datetime(dados_hanseniase_ano_mes['Data'])
```

```python
plt.figure(figsize=(22,12))
ax = sns.lineplot(x='Data', y='Casos', data=dados_hanseniase_ano_mes,linewidth=2.5,label='Casos')
plt.title('Casos de Hanseniase ao longo dos anos', Fontsize=20, pad=10)
ax.xaxis.set_major_locator(ticker.MaxNLocator((len(dados_hanseniase_ano_mes)/6)))
plt.xticks(rotation=60)
plt.legend()
plt.grid()
plt.show()
```

Fim do Desafio

```python
dados_st['Media_movel_3'] = dados_st['Casos'].rolling(3).mean()
```

```python
plt.figure(figsize=(22,12))
ax = sns.lineplot(x='Data', y='Casos', data=dados_st,linewidth=2.5,label='Casos')
sns.lineplot(x='Data', y='Media_movel_3', data=dados_st,linewidth=2.5, color='red',label='Media movel 3 meses')
plt.title('Casos de Tuberculose ao longo dos anos', Fontsize=20, pad=10)
ax.xaxis.set_major_locator(ticker.MaxNLocator((len(dados_st)/6)))
plt.xticks(rotation=60)
plt.legend()
plt.grid()
plt.show()
```

```python
dados_st['Media_movel_6'] = dados_st['Casos'].rolling(6).mean()
dados_st['Media_movel_12'] = dados_st['Casos'].rolling(12).mean()
```

```python
plt.figure(figsize=(22,12))
ax = sns.lineplot(x='Data', y='Casos', data=dados_st,linewidth=2.5,label='Casos')
sns.lineplot(x='Data', y='Media_movel_3', data=dados_st,linewidth=2.5, color='green',label='Media movel 3 meses')
sns.lineplot(x='Data', y='Media_movel_6', data=dados_st,linewidth=2.5, color='yellow',label='Media movel 6 meses')
sns.lineplot(x='Data', y='Media_movel_12', data=dados_st,linewidth=2.5, color='red',label='Media movel 12 meses')
plt.title('Casos de Tuberculose ao longo dos anos', Fontsize=20, pad=10)
ax.xaxis.set_major_locator(ticker.MaxNLocator((len(dados_st)/6)))
plt.xticks(rotation=60)
plt.legend()
plt.grid()
plt.show()
```

```python
# Data set esta fora de ordem
```

```python
dados_st = dados_st.sort_values(by=['Data'])
dados_st.reset_index(drop=True,inplace=True)
dados_st['Media_movel_3'] = dados_st['Casos'].rolling(3).mean()
dados_st['Media_movel_6'] = dados_st['Casos'].rolling(6).mean()
dados_st['Media_movel_12'] = dados_st['Casos'].rolling(12).mean()
```

```python
plt.figure(figsize=(22,12))
ax = sns.lineplot(x='Data', y='Casos', data=dados_st,linewidth=1.5,label='Casos')
sns.lineplot(x='Data', y='Media_movel_3', data=dados_st,linewidth=1.5, color='green',label='Media movel 3 meses')
sns.lineplot(x='Data', y='Media_movel_6', data=dados_st,linewidth=1.5, color='yellow',label='Media movel 6 meses')
sns.lineplot(x='Data', y='Media_movel_12', data=dados_st,linewidth=1.5, color='red',label='Media movel 12 meses')
plt.title('Casos de Tuberculose ao longo dos anos', Fontsize=20, pad=10)
ax.xaxis.set_major_locator(ticker.MaxNLocator((len(dados_st)/6)))
plt.xticks(rotation=60)
plt.legend()
plt.grid()
plt.show()
```

```python
dados_st['Media_movel_3c'] = dados_st['Casos'].rolling(3,center=True).mean()
```

```python
from statsmodels.tsa.arima_model import ARIMA

modelo_arima = ARIMA(dados_st['Casos'].values.astype('float32'), order=[0,0,0])
modelo_arima_treinado = modelo_arima.fit()
```

```python
modelo_arima_treinado.forecast(12)
```

## Desafio 09 - Na tupla de retorno do forecast, pesquisar o que sao os elementos

O item 2 é o desvio padrão  
E o ultimo é o intervalo de confiança  


Fim do Desafio

```python
modelo_arima_treinado.forecast(12)[0]
```

```python
dados_treino = dados_st.query("Ano != 2019")
dados_teste = dados_st.query("Ano == 2019")
modelo_arima = ARIMA(dados_treino['Casos'].values.astype('float32'), order=[0,0,0])
modelo_arima_treinado = modelo_arima.fit()
```

```python
previsao = modelo_arima_treinado.forecast(12)[0]
```

```python
sns.lineplot(x='Data', y='Casos', data=dados_treino)
sns.lineplot(x='Data', y='Casos', data=dados_teste)
```

```python
dados_teste['Previsao'] = previsao
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(x='Data', y='Casos', data=dados_treino)
sns.lineplot(x='Data', y='Casos', data=dados_teste)
sns.lineplot(x='Data', y='Previsao', data=dados_teste)
```

```python
modelo_arima = ARIMA(dados_treino['Casos'].values.astype('float32'), order=[12,1,0])
modelo_arima_treinado = modelo_arima.fit()
previsao = modelo_arima_treinado.forecast(12)[0]
dados_teste['Previsao'] = previsao
```

```python
plt.figure(figsize=(18,10))
sns.lineplot(x='Data', y='Casos', data=dados_treino)
sns.lineplot(x='Data', y='Casos', data=dados_teste)
sns.lineplot(x='Data', y='Previsao', data=dados_teste)
```

```python
modelo_arima = ARIMA(dados_treino['Casos'].values.astype('float32'), order=[12,1,2])
modelo_arima_treinado = modelo_arima.fit()
previsao = modelo_arima_treinado.forecast(12)[0]
dados_teste['Previsao'] = previsao
plt.figure(figsize=(18,10))
sns.lineplot(x='Data', y='Casos', data=dados_treino)
sns.lineplot(x='Data', y='Casos', data=dados_teste)
sns.lineplot(x='Data', y='Previsao', data=dados_teste)
```

## Desafio 10 - Variar os parametros max_p, max_d e max_q

```python
modelo_autoarima_teste = modelo_autoarima = auto_arima(dados_treino['Casos'].values, error_action='ignore',trace=True,max_D=120,max_P=120,max_q=120,m=12)
```

```python
modelo_autoarima_teste.predict(12)
```

```python
previsao_autoarima = modelo_autoarima_teste.predict(12)
dados_teste['Previsao_autoarima'] = previsao_autoarima
plt.figure(figsize=(18,10))
sns.lineplot(x='Data', y='Casos', data=dados_treino)
sns.lineplot(x='Data', y='Casos', data=dados_teste)
sns.lineplot(x='Data', y='Previsao_autoarima', data=dados_teste)
```

Fim do Desafio

```python
modelo_arima = ARIMA(dados_treino['Casos'].values.astype('float32'), order=[12,1,2])
modelo_arima_treinado = modelo_arima.fit()
previsao_autoarima = modelo_autoarima.predict(12)
dados_teste['Previsao_autoarima'] = previsao_autoarima
plt.figure(figsize=(18,10))
sns.lineplot(x='Data', y='Casos', data=dados_treino)
sns.lineplot(x='Data', y='Casos', data=dados_teste)
sns.lineplot(x='Data', y='Previsao_autoarima', data=dados_teste)
```

```python
# Incluindo sasonalidade
modelo_autoarima = auto_arima(dados_treino["Casos"].values, error_action="ignore", trace=True, seasonal=True, m = 12)
```

```python
modelo_arima = ARIMA(dados_treino['Casos'].values.astype('float32'), order=[12,1,2])
modelo_arima_treinado = modelo_arima.fit()
previsao_autoarima = modelo_autoarima.predict(12)
dados_teste['Previsao_autoarima'] = previsao_autoarima
plt.figure(figsize=(18,10))
sns.lineplot(x='Data', y='Casos', data=dados_treino)
sns.lineplot(x='Data', y='Casos', data=dados_teste)
sns.lineplot(x='Data', y='Previsao_autoarima', data=dados_teste)
```

```python
from sklearn.metrics import mean_squared_error

resultado_autoarima = mean_squared_error(dados_teste["Casos"], dados_teste["Previsao_autoarima"])
resultado_arima = mean_squared_error(dados_teste["Casos"], dados_teste["Previsao"])
print(f"Autoarima = {resultado_autoarima}")
print(f"Arima = {resultado_arima}")
```
