## Desafio-1 - Transformar dados em uma serie ano-mes

```python
dados_st = dados.melt(id_vars='Ano Diagn√≥stico', value_name='Casos', var_name='M√™s')
```

```python
dados_st['Data'] = dados_st['Ano Diagn√≥stico']+'-'+dados_st['M√™s']
dados_st
```

Fim do Desafio

```python
plt.figure(figsize=(18,12))
sns.lineplot(x='Data', y='Casos', data=dados_st[:30])
```

## Desafio-2 - Converter a data para Datetime

```python
#pd.to_datetime(dados_st['Data'],yearfirst=True)

#Erro pois os nomes estao em portugues
```

```python
dic_mes = {
    'Jan':'Jan',
    'Fev':'Feb',
    'Mar':'Mar',
    'Abr':'Apr',
    'Mai':'May',
    'Jun':'Jun',
    'Jul':'Jul',
    'Ago':'Aug',
    'Set':'Sep',
    'Out':'Oct',
    'Nov':'Nov',
    'Dez':'Dec',
}
dados_st['M√™s'] = dados_st['M√™s'].map(dic_mes)
dados_st['Data'] = dados_st['Ano Diagn√≥stico']+'-'+dados_st['M√™s']
pd.to_datetime(dados_st['Data'])
```

```python
dados_st
```

 Fim do Desafio

```python
dados_st = dados_st[['Data','Casos']]
dados_st
```

```python
plt.figure(figsize=(22,12))
sns.lineplot(x='Data', y='Casos', data=dados_st)
plt.xticks(rotation=90)
plt.show()
```
